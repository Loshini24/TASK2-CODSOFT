```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.figure_factory as ff
import plotly.offline as py
```

```python
file = '../input/imdb-india-movies/IMDb Movies India.csv'
import chardet
with open(file, 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
result
```

```python
df = pd.read_csv(file,encoding='ISO-8859-1')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

Finding out the number on null elements in each column

```python
df.isnull().sum()
```

Whoa! thats a lot of null elements... let's clean this shit

```python
df1 = df.dropna(axis=0, subset=['Year'])  #dropping rows where Year is missing
```

<h1>Finding out the director with most movies</h1>

https://www.kaggle.com/aliendev/example-of-pandas-dropna

```python
director = df1.dropna(axis=0, subset=['Director'])
```

```python
dfg = director['Director'].value_counts().head(20)
```

```python
fig = px.bar(dfg)
fig.show()
```

<h1>Finding out actor with most movies</h1>

```python
actor = df1.dropna(axis=0, subset=['Actor 1'])
```

```python
act = actor['Actor 1'].value_counts().head(20)
```

```python
fig = px.bar(act)
fig.show()
```

<h1>Finding out top movie of each year</h1>

```python
rating = df1.dropna(axis=0, subset=['Rating'])  #dropping rows where Year and rating is missing
```

```python
rating = rating.sort_values('Rating', ascending=False).drop_duplicates(['Year'])
```

```python
data = ff.create_table(rating)
py.iplot(data)
```

<h1>Analysing the Genre Distribution</h1>

```python
df2 = df1.dropna(axis=0, subset=['Genre'])  #dropping rows where genre is missing
```

```python
gen = df2['Genre'].value_counts()
```

```python
gen
```

```python
dfc = pd.DataFrame(list(gen.items()))
dfc = dfc.rename(columns={0: 'Genre', 1: 'Count'})
dfc
```

```python
dfc.loc[dfc['Count'] < 50, 'Genre'] = 'Other' # Represent only certain genres
fig = px.pie(dfc, values='Count', names='Genre', title='Genre Distribution in Indian Cinema')
fig.show()
```
